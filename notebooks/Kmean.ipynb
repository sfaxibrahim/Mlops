```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Assuming 'df_numeric' is your prepared data
df_numeric = data.drop(['timestamp', 'Unnamed: 0'], axis=1)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=23)
kmeans.fit(df_numeric)

# Get the cluster labels
labels = kmeans.labels_

# Optional: Reduce the dimensionality to 2D for visualization using PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_numeric)

# Plot the clusters in 2D space
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels, cmap='viridis', marker='o')
plt.title("K-Means Clustering with 3 Clusters")
plt.grid(True)
plt.show()
```
